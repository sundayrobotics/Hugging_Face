Audio File to Text

```python
import librosa #manages audio files
import torch # importing pytorch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
import IPython.display as display
file = display.Audio("file1.wav", autoplay=False)
file
```

```python
# asudio files have to be 16kHz sampling rate
audio, rate = librosa.load('file1.wav', sr=16000)
```

```python
audio #audio is float32
```

```python
rate
```

```python
#import the pre-trained model

tokenizer = Wav2Vec2Tokenizer.from_pretrained('facebook/wav2vec2-base-960h')
model = Wav2Vec2ForCTC.from_pretrained('facebook/wav2vec2-base-960h')
```

```python
input_values = tokenizer(audio, return_tensors='pt').input_values
input_values
```

```python
# get and store logits
logits = model(input_values).logits
logits
```

```python
# pass the logits to softmax to get the predicted values
prediction = torch.argmax(logits, dim=-1)
```

```python
# convert the audio to text
transcription = tokenizer.batch_decode(prediction)[0]
transcription.capitalize()
```
